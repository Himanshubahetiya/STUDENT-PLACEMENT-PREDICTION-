```python
import pandas as pd
import pickle
```

```python
with open("stp.pkl","rb")as f:
    model = pickle.load(f)
```

```python
age = int(input("Enter age: "))
gender = input("Enter gender (Male/Female): ")
branch = input("Enter branch (CSE/IT/ECE/Mechanical): ")
cgpa = float(input("Enter CGPA: "))
internship_count = int(input("Enter internship count: "))
project_count = int(input("Enter project count: "))
certifications_count = int(input("Enter certifications count: "))
coding_skills_score = int(input("Enter coding skills score: "))
communication_skills_score = int(input("Enter communication skills score: "))
soft_skills_score = int(input("Enter soft skills score: "))
# hackathon_participation = input("Hackathon participation (Yes/No): ")
```

```python
data = pd.DataFrame([{
    "age": age,
    "gender": gender,
    "branch": branch,
    "cgpa": cgpa,
    "internship_count": internship_count,
    "project_count": project_count,
    "certifications_count": certifications_count,
    "coding_skills_score": coding_skills_score,
    "communication_skills_score": communication_skills_score,
    "soft_skills_score": soft_skills_score
}])
```

```python
data
```

```python
prediction = model.predict(data)

print("Prediction:", prediction[0])
```
