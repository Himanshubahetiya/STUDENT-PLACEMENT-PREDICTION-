```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report
```

```python
data = pd.read_csv("student_placement_prediction_dataset_2026.csv")
```

```python
data
```

```python
data.isnull().sum()
```

```python
df = data.drop(["salary_package_lpa", "student_id", "linkedin_connections","extracurricular_score", "sleep_hours","volunteer_experience", "attendance_percentage", "study_hours_per_day"], axis=1)
```

```python
important_features = [
"mock_interview_score",
"aptitude_score",
"coding_skill_score",
"communication_skill_score",
"logical_reasoning_score",
"leadership_score",
"cgpa",
"backlogs",
"projects_count",
"internships_count"
]
```

```python
x = df[important_features]
```

```python
df.columns
```

```python
y = df["placement_status"]
# x = df.drop("placement_status", axis=1)
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x,y, random_state = 42, 
)
```

```python
cat_cols = x.select_dtypes(include = "object").columns
num_cols = x.select_dtypes(exclude="object").columns
```

```python
preprocessor = ColumnTransformer(
    transformers=[
        ("cat", OneHotEncoder(handle_unknown="ignore"), cat_cols),
        ("num", "passthrough", num_cols)
    ]
)
```

```python
pipeline = Pipeline([
    ("preprocessing", preprocessor),
    ("model", DecisionTreeClassifier(
        max_depth=12,
        min_samples_split=10,
        min_samples_leaf=5,
        class_weight="balanced",
        random_state=42
    ))
])
```

```python
pipeline.fit(x_train, y_train)
```

```python
y_pred = pipeline.predict(x_test)
```

```python
print("Accuracy:", accuracy_score(y_test, y_pred))
print("classification_report", classification_report(y_test, y_pred))
```

```python
feature_names = pipeline.named_steps["preprocessing"].get_feature_names_out()

importances = pipeline.named_steps["model"].feature_importances_

import pandas as pd

importance_df = pd.DataFrame({
    "feature": feature_names,
    "importance": importances
}).sort_values("importance", ascending=False)

print(importance_df.head(10))
```
